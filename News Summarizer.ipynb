```python
import bs4 
import html5lib
import requests
```

```python
response= requests.get(r"https://www.vox.com/policy-and-politics/2019/5/30/18642535/california-ab5-misclassify-employees-contractors")
```

```python
type(response.text)
```

```python
type(response.content)
```

```python
content=response.content
```

```python
soup=bs4.BeautifulSoup(content)
```

```python
soup
```

```python
paras=soup.find_all('p')
```

```python
for para in paras:
    print(para.text,"/n/n***********************************************************************************************************************************************************")
```

```python
news=''
for para in paras:
    news += para.text
    
    
```

```python
news
```

```python
title=soup.find_all('title')[0].text
```

```python
title
```

```python
len(news.split())
```

```python
import gensim
```

```python
from gensim.summarization.summarizer import summarize
from gensim.summarization import keywords
```

```python
summary=summarize(news, word_count=500)
```

```python
print(f'A 500 words summary of the news article titled {title} is:\n')
print(summary)
```

```python
len(summary.split())
```

```python
print("Keywords pertaining to this news article are: \n ")
print(keywords(news, ratio=0.1, lemmatize=True))
```
